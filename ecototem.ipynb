```python
df7 = df1.resample('2H').sum()
```

```python
# 
trace1 = go.Scatter(
    x = df7.index,
    y = df7.Total,
    mode = 'lines',
    line=Line(
        color='orange',
        shape='spline'
    ),
    name = 'both directions',
    visible=True,
)
#total count, working days
trace2 = go.Scatter(
    x = df7.index,
    y = df7.Westbound,
    mode = 'lines',
    line=Line(
        color='green',
        shape='spline'
    ),
    name = 'Westbound',
    visible=False,
)
#total count, all days
trace3 = go.Scatter(
    x = df7.index,
    y = df7.Eastbound,
    mode = 'lines',
    line=Line(
        color='red',
        shape='spline'
    ),
    name = 'Eastbound',
    visible=False,
)

data = Data([trace1, trace2, trace3])

#BUTTONS
layout = Layout(
    title='Ecototem — counting cyclists over time',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1 days',
                     step='day',
                     stepmode='backward'),
                dict(count=7,
                     label='7 days',
                     step='day',
                     stepmode='forward'),
                dict(count=1,
                     label='30 days',
                     step='month',
                     stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(),
        type='date'
    ),
    updatemenus=list([
        dict(
            x=-0.05,
            y=1,
            yanchor='top',
            buttons=list([
                dict(
                    args=['visible', [True, False, False]],
                    label='Both direction',
                    method='restyle'
                ),
                dict(
                    args=['visible', [False,True, False]],
                    label='To Westbound',
                    method='restyle'
                ),
                dict(
                    args=['visible', [False, False, True]],
                    label='To Eastbound',
                    method='restyle'
                ),
                dict(
                    args=['visible', [True, True, True]],
                    label='All',
                    method='restyle'
                )
            ]),
        )
    ]),
)






fig = Figure(data=data, layout=layout)
iplot(fig)
```
