```python
HW = pd.read_csv('data/Trips/allTrips.csv')
```

```python
HW.columns
```

```python
date = pd.to_datetime(arg=HW['starttime'])
df = pd.concat([date.dt.month, HW.tripduration], axis=1)
df.columns = ['month','duration']
df = df[df.duration < 0.1*3600]
df = df[df.duration != 0]
df.sort_values(by='month', inplace=True)
```

```python
hist = Histogram(df, values='duration',color='month')
show(hist)
```
