```python
%matplotlib inline
#import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
import plotly.plotly as py
# plotly.tools.set_credentials_file(username='marced', api_key='7meNCtLRU4D2vUuHqCdP')
import plotly.graph_objs as go
from plotly.graph_objs import *
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
import plotly.tools as tls
import datetime
from bokeh.charts import output_file, Chord
from bokeh.io import show
from bokeh.sampledata.les_mis import data
import math
```

```python
df = pd.read_csv('data/GoogleMapsScrap/frequency.csv', index_col=[0])
```

```python
origin = []
destination = []
frequency = []
x_o = []
y_o = []
x_d = []
y_d = []
for o in range(0,len(df)):
    for d in range(0,len(df)):
        origin.append(df.index[o])
        destination.append(df.columns[d])
        frequency.append(df.iloc[o,d])
        theta_o = 359/(len(df))*o
        theta_d = 359/(len(df))*d
        x_o.append(50* math.cos(theta_o))
        y_o.append(50* math.sin(theta_o))
        x_d.append(50* math.cos(theta_d))
        y_d.append(50* math.sin(theta_d))
```

```python
df1 = pd.DataFrame([origin, destination, frequency,x_o,y_o,x_d,y_d])
df1 = df1.T
df1.columns = ['origin', 'destination', 'frequency','x_origin','y_origin','x_destination','y_destination']
df1 = df1[df1.frequency > 0]
df1 = df1[df1.origin != df1.destination]
df1 = df1[df1.frequency > 100]
df1 = df1.reset_index()
#df1 = df1[np.isfinite[df1.frequency]]
```

```python
coordinates = df1[['origin','x_origin', 'y_origin']]
coordinates = coordinates.drop_duplicates()
```

```python
#traces0 = []
traces_all = []

df_all = df1
#df_300 = df1[df1.frequency > 300].reset_index()
#df_1000 = df1[df1.frequency > 1000].reset_index()

traces_all.append(go.Scatter(x = coordinates.x_origin,
                     y = coordinates.y_origin,
                     text = coordinates.origin,
                    hoverinfo='text',
                    mode='markers',
                             opacity = 0.5,
           marker=Marker(symbol='dot',
                         size=4,
                         color='orange'
                         )))


for i in range(0,len(df_all[:])):
    traces_all.append(go.Scatter(x = [df_all.x_origin[i], df_all.x_destination[i]],
                                y = [df_all.y_origin[i], df_all.y_destination[i]], 
                                mode = 'lines',
                             opacity = 0.1,
                             hoverinfo='none',
                             line=Line(width = df_all.frequency[i]/1000,
                                        shape='spline',
                                      color = 'orange')))
data = Data(traces_all)

axis=dict(showline=False, # hide axis line, grid, ticklabels and  title
          zeroline=False,
          showgrid=False,
          showticklabels=False,
          title=''
          )

width=700
height=750

layout=Layout(title= 'All links',
              font= Font(size=12),
              showlegend=False,
              hovermode='closea',
              autosize=False,
              width=width,
              height=height,
              xaxis=XAxis(axis),
              yaxis=YAxis(axis),
              margin=Margin(l=40,
                            r=40,
                            b=85,
                            t=100,
                          )
             )

fig= go.Figure(data=data, layout=layout)
iplot(fig)
```
