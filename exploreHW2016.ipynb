```python
#HW16 = pd.concat([HW16, d], axis = 1)
HW11_14 = pd.read_csv('data/Trips/allTrips11_14.csv', parse_dates=True, index_col=[0])
HW15_17 = pd.read_csv('data/Trips/allTrips.csv', parse_dates=True, index_col=[0])
weather = pd.read_csv('data/weather/weatherCambridge_python2.csv', delimiter=';', parse_dates=True)
```

```python
grouped15_17 = HW15_17.resample('24H').count()
grouped11_14 = HW11_14.resample('24H').count()
date15_17 = grouped15_17.index
```

```python
counter11_14 = grouped11_14.Gender
counter15_17 = grouped15_17.gender
date = pd.concat([counter11_14,counter15_17])
date = date.reset_index()
date.columns = ['date','counter']
```

```python
trace1 = go.Bar(
    x=date.date,
    y=date.counter,
    opacity=0.75,
    name='#trips',
    marker=dict(
        color='orange',
        line=dict(
            color='orange',
        )
    ),
    visible=True,
)

trace2 = go.Scatter(
    x=date15_17,
    y=weather.temp_avg,
    fill= None,
    mode='markers',
    yaxis = 'y2',
    name='Average temperature',
    marker=dict(
        size='6',
        color = weather.temp_avg, #set color equal to a variable
        colorscale = [[0, '#254C72'],[0.5, '#E4B21B'], [1, '#AF3030']],
        #colorscale='Viridis',
        showscale=False,
        opacity=0.6,
    ),
    visible=False,
)

trace3 = go.Bar(
    x=date15_17,
    y=weather.precip,
    opacity=0.75,
    name='#trips',
    marker=dict(
        color='orange',
        line=dict(
            color='orange',
        )
    ),
    visible=False,
)

trace4 = go.Bar(
    x=date15_17,
    y=weather.wind_avg,
    opacity=0.75,
    yaxis = 'y4',
    name='#trips',
    marker=dict(
        color='red',
        line=dict(
            color='red',
        )
    ),
    visible=False,
)

trace5 = go.Scatter(
    x=date15_17,
    y=weather.hum_avg,
    fill= None,
    mode='markers',
    yaxis = 'y3',
    name='Average temperature',
    marker=dict(
        size='6',
        color = weather.temp_avg, #set color equal to a variable
        colorscale = [[0, '#A8A9B2'],[0.5, '#86B1E4'], [1, '#2D43CB']],
        #colorscale='Viridis',
        showscale=False,
        opacity=0.6,
    ),
    visible=False,
)
data = [trace2, trace3, trace4, trace5, trace1]

updatemenus = list([
    dict(type="buttons",
         active=0,
         buttons=list([   
            dict(label = '#trips',
                 method = 'restyle',
                 args = [{'visible': [False, False, False,False, True]},
                         ]),
            dict(label = 'Avg Hum',
                 method = 'restyle',
                 args = [{'visible': [False, False, False, True, True]},
                         ]),
            dict(label = 'Trips vs Temp',
                 method = 'update',
                 args = [{'visible': [True, False, False,False, True]},
                         ]),
            dict(label = 'Wind',
                 method = 'restyle',
                 args = [{'visible': [False, False, True,False, True]},
                         ])
        ]),
    )
])

layout = go.Layout(
    #showlegend=False,
    title='Exploring Weather vs trips correlation',
    updatemenus=updatemenus,
    xaxis=dict(
        domain = [0.08,0.8]
    ),
    yaxis=dict(
        title='# trips'
    ),
    yaxis2=dict(
        title='Temperature °C',
        overlaying='y',
        side='right',
        showgrid=False,
        zeroline=False,
        showline=False,
    ),
    yaxis3=dict(
        title='Humidity',
        overlaying='y',
        side='right',
        autorange='reversed',
        showgrid=False,
        zeroline=False,
        showline=False,
        anchor='free',
        position=0.9
    ),
    yaxis4=dict(
        title='Wind',
        overlaying='y',
        side='left',
        position=0,
        anchor='free',
        autorange='reversed',
        showgrid=False,
        zeroline=False,
        showline=False,
    ))
fig = go.Figure(data=data, layout=layout)

iplot(fig, filename='temperature')
```
