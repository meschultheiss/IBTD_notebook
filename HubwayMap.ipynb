# With flows

```python
freq = pd.read_csv('data/GoogleMapsScrap/frequency.csv', index_col=[0])
location = pd.read_csv('data/GoogleMapsScrap/repository.csv')
```

```python
occurence = location.occurence
occurence_scale = occurence/10000
for i in range(0,len(occurence)):
    if occurence_scale.iloc[i] < 4:
        occurence_scale.iloc[i] = 4
```

```python
occurence_perc = occurence*1000/occurence.sum()
occurence_perc = occurence_perc.astype(int)
```

```python
location.head(6)
```

```python
lat_line = []
lon_line = []
intensity = []
i = 0
j = 0
for i in range(0, len(location.station_name)):
    for j in range(0, len(location.station_name)):
        o = np.where(freq.index == location.station_name[i])[0][0]
        d = np.where(freq.columns == location.station_name[j])[0][0]
        if freq.iloc[o,d] > 300:
            lat_line.append(location.latitude[i])
            lat_line.append(location.latitude[j])
            lon_line.append(location.longitude[i])
            lon_line.append(location.longitude[j])
            intensity.append(freq.iloc[o,d])
```

```python
intensity_map = []
for i in intensity:
    intensity_map.append(i/20000)
```

```python
mapbox_access_token = 'pk.eyJ1IjoibWFyY2VkIiwiYSI6ImNqMGNtZjk4cjA0NWMzMm12ZjZ4ejEzcWkifQ.gdCtrZod8K7UnClAkUJ0rg'

data1 = Scattermapbox(
        lat=lat_line,
        lon=lon_line,
        mode='lines',
        line=Line(
            color = 'orange',
            width=intensity_map,
        ),
        opacity= 0.035,
        hoverinfo='text'
    )

data2 = Scattermapbox(
        lat=location.latitude,
        lon=location.longitude,
        mode='markers',
        marker=Marker(
            size=occurence_scale,
            color='orange',
            opacity=0.8,
        ),
        text= location.station_name +"  Occurence: "+ occurence_perc.astype('str')  + " of all the trips",
        hoverinfo='text'
    )

data = Data([data2,data1])

layout = Layout(
    title='Hubway map',
    autosize=True,
    hovermode='closest',
    showlegend=False,
    width=1000,
    height=700,
    mapbox=dict(
        accesstoken=mapbox_access_token,
        bearing=0,
        center=dict(
            lat=42.358118,
            lon=-71.09
        ),
        pitch=0,
        zoom=11,
        style='mapbox://styles/marced/cj0cmj7zr003y2rqhdu4pzrtj'
    ),
)



fig = dict(data=[data1,data2], layout=layout)
#iplot(fig, filename='HWmap')
iplot(fig, filename='HWmap',image='png', image_height=1000, image_width=1200)
```
